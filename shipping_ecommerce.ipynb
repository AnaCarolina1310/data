DATA VISUALIZATION

```python
class_by_product_importance= df.groupby(['Product_importance'])['Class'].value_counts().reset_index() 
class_by_product_importance['%']=round((class_by_product_importance['count'] / class_by_product_importance['count'].sum()) * 100, 2)
class_by_product_importance['Rate']= round(class_by_product_importance['count']/class_by_product_importance['count'].sum(),2)
class_by_product_importance
```

```python
class_by_product_importance_bar= sns.barplot(x='Product_importance', hue='Class',y='count', data=class_by_product_importance)
```

La tasa de entregas a tiempo supera a la tasa de envios con retraso en todas las categorías de importancia de los productos. Es mayor la cantidad de productos que se envían con importancia entre media y baja. 

```python
class_by_mode_of_shipment= df.groupby(['Mode_of_Shipment'])['Class'].value_counts().reset_index() 
class_by_mode_of_shipment['%']=round((class_by_mode_of_shipment['count'] / class_by_mode_of_shipment['count'].sum()) * 100, 2)
class_by_mode_of_shipment['Rate']= round(class_by_mode_of_shipment['count']/class_by_mode_of_shipment['count'].sum(),2)
class_by_mode_of_shipment
```

```python
class_by_mode_of_shipment_bar= sns.barplot(x='Mode_of_Shipment', hue='Class',y='count', data=class_by_mode_of_shipment)
```

Mas del 60% de los envíos se trasladan por barco, el cual tiene una tasa de retraso del 59%. Los envíos por ruta y los aéreos tienen un 40% y 39% de retraso respectivamente, siempre teniendo en cuenta el total de envíos que se realizan por cada vía.

```python
dicount_by_class=sns.boxplot(y='Class',x="Discount_offered", data=df)
dicount_by_class.set(title="Dispersion of discount by shipment status",xlabel='Discount',ylabel='Shipment Status')
```

El 50% de los envíos con retraso tuvieron un descuento de un 8% y un máximo de 10%. En cambio, en los envíos que llegaron en termino se observa mayor dispersión, donde el 50% de los envíos tuvieron un descuento que ronda el 10% y un máximo mayor al 60%.

```python
class_by_warehouse= df.groupby(['Warehouse_block'])['Class'].value_counts().reset_index() 
class_by_warehouse['%']=round((class_by_warehouse['count'] / class_by_warehouse['count'].sum()) * 100, 2)
class_by_warehouse['Rate']= round(class_by_warehouse['count']/class_by_warehouse['count'].sum(),2)
class_by_warehouse
```

```python
class_by_warehouse_bar= sns.barplot(y='Warehouse_block', hue='Class',x='count', data=class_by_warehouse)
```

Todos los warehouses blocks tienen la misma cantidad de envíos totales, a excepción del F. Este último tiene el doble de envíos que los demás pero en termino de tasa de retrasos todos son igualmente proporcionales.

```python
weight_by_class=sns.boxplot(y='Weight_in_gms',x="Class", data=df)
dicount_by_class.set(title="Dispersion of discount by shipment status",xlabel='Shipment Status',ylabel='Discount')
```

La distribución del peso de los envíos entregados a tiempo es mayor que los que llegan con retraso. Se podría deducir que mientras más pesado sea el producto a enviar hay más posibilidades de retraso.
