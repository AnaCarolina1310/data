Descripción:
- Hotel : Hotel ( Resort Hotel o City Hotel)
- Is_canceled : Valor que indica si la reserva fue cancelada (1) o no (0) 
- Lead_time : Número de días transcurridos entre la fecha de entrada de la reserva en el PMS y la fecha de llegada
- Arrival_date_year : Año de la fecha de llegada
- Arrival_date_month : Mes de la fecha de llegada
- Arrival_date_week_number : Número de semana del año de la fecha de llegada
- Arrival_date_day_of_month : Día de la fecha de llegada
- Stays_in_weekend_nights : Número de noches de fin de semana (sábado o domingo) que el huésped se alojó o reservó para alojarse en el hotel
- Stays_in_week_nights : Número de noches entre semana (de lunes a viernes) que el cliente se alojó o reservó en el hotel
- Adults : Número de adultos
- Niños : Número de niños
- Babies : Número de bebés
- Comida : Tipo de comida reservada. Las categorías se presentan en paquetes de comidas de hospitalidad estándar: Sin definir/SC sin paquete de comidas; BB - Alojamiento y desayuno; HB - Media pensión (desayuno y otra comida - normalmente cena); FB - Pensión completa (desayuno, almuerzo y cena).
- País : País de origen. Las categorías se representan en el formato ISO 3155-3:2013
- Segmento_mercado : Designación del segmento de mercado. En las categorías, el término "TA" significa "agencias de viajes" y "TO" significa "tour operadores".
- Distribution_channel : Canal de distribución de las reservas. El término "TA" significa "Agencias de viajes" y "TO" significa "Tour operadores".
- Is_repeated_guest : Valor que indica si el nombre de la reserva era de un huésped repetido (1) o no (0)
- Previous_cancellations : Número de reservas anteriores canceladas por el cliente antes de la reserva actual
- Previous_bookings_not_canceled : Número de reservas anteriores no canceladas por el cliente antes de la reserva actual
- Reserved_room_type : Código del tipo de habitación reservada. El código se presenta en lugar de la designación por razones de anonimato.
- Tipo_habitación_asignada : Código del tipo de habitación asignado a la reserva. A veces, el tipo de habitación asignada difiere del tipo de habitación reservada por motivos de funcionamiento del hotel (por ejemplo, exceso de reservas) o a petición del cliente. El código se presenta en lugar de la designación por razones de anonimato.
- Booking_changes : Número de cambios/enmiendas realizados en la reserva desde el momento en que se introdujo la reserva en el PMS hasta el momento de la facturación o la cancelación.
- Deposit_type : Indicación de si el cliente ha efectuado un depósito para garantizar la reserva. Esta variable puede ser de tres tipos
- Agent : ID de la agencia de viajes que efectuó la reserva
- Empresa : ID de la empresa/entidad que realizó la reserva o responsable del pago de la reserva. Se presenta el ID en lugar de la designación por razones de anonimato
- Days_in_waiting_list : Número de días que la reserva estuvo en lista de espera antes de ser confirmada al cliente
- Customer_type : Tipo de reserva, asumiendo una de las cuatro categorías: Contract - cuando la reserva tiene asociada una adjudicación u otro tipo de contrato; Group - cuando la reserva está asociada a un grupo; Transient - cuando la reserva no forma parte de un grupo o contrato, y no está asociada a otra reserva transient; Transient-party - cuando la reserva es transient, pero está asociada al menos a otra reserva transient
- Adr : Tarifa media diaria definida dividiendo la suma de todas las transacciones de alojamiento por el número total de noches de estancia
- Required_car_parking_spaces : Número de plazas de aparcamiento requeridas por el cliente
- Total_of_special_requests : Número de peticiones especiales realizadas por el cliente (por ejemplo, cama doble o piso alto)
- Reservation_status_date : Fecha del último estado de la reserva.

## a) Análisis exploratorio


V. Hotel

```python
train['hotel'].value_counts().plot.barh(alpha=0.5, title="Cantidad de reservas por hotel", xlabel='Frecuencia',ylabel='Hotel') 
```

```python
v_hotel=snb.barplot(x='hotel', y= 'is_canceled', data=train)
v_hotel.set_title('Tasa de cancelaciones por hotel')
v_hotel.set_xlabel('Mes de arribo')
v_hotel.set_ylabel('Tasa de cancelación')
```

```python
canceled_by_hotel= train.groupby('hotel')['is_canceled'].sum().reset_index()
v_hotel_cant=snb.barplot(x='hotel', y= 'is_canceled', data=canceled_by_hotel)
v_hotel_cant.set_title('Cantidad de cancelaciones por hotel')
v_hotel_cant.set_xlabel('Hotel')
v_hotel_cant.set_ylabel('Frecuencia')
```

Mas del 60% de las reservas estudiadas son de City Hoteles. Las reservas canceladas son mayores en la misma categoría, con una tasa de cancelacion superior al 0.5.

V. Mes de llegada

```python
train['arrival_date_month'].value_counts()
```

```python
bar_hotel=train['arrival_date_month'].value_counts().plot.bar(alpha=0.5 , title="Cantidad de reservas por hotel ", xlabel="Hotel",ylabel="Frecuencia")
```

```python
bar_cancel_arrival_month=snb.barplot(x='arrival_date_month', y= 'is_canceled', data=train)
bar_cancel_arrival_month.set_xticklabels(bar_cancel_arrival_month.get_xticklabels(), rotation=45)
bar_cancel_arrival_month.set_title('Tasa de cancelacion por mes de arribo')
bar_cancel_arrival_month.set_xlabel('Mes de arribo')
bar_cancel_arrival_month.set_ylabel('Tasa de cancelación')
```

```python
bar_cancel_arrival_month_by_hotel=snb.barplot(x='arrival_date_month', y= 'is_canceled',hue='arrival_date_year',data=train)
bar_cancel_arrival_month_by_hotel.set_xticklabels(bar_cancel_arrival_month.get_xticklabels(), rotation=45)
bar_cancel_arrival_month_by_hotel.set_title('Tasa de cancelacion por mes de arribo y por año')
bar_cancel_arrival_month_by_hotel.set_xlabel('Mes de arribo')
bar_cancel_arrival_month_by_hotel.set_ylabel('Tasa de cancelación')
```

```python
bar_cancel_arrival_month_by_hotel=snb.barplot(x='arrival_date_month', y= 'is_canceled',hue='hotel',data=train)
bar_cancel_arrival_month_by_hotel.set_xticklabels(bar_cancel_arrival_month.get_xticklabels(), rotation=45)
bar_cancel_arrival_month_by_hotel.set_title('Tasa de cancelacion por mes de arribo y por hotel')
bar_cancel_arrival_month_by_hotel.set_xlabel('Mes de arribo')
bar_cancel_arrival_month_by_hotel.set_ylabel('Tasa de cancelación')
```

El periodo de mayor cantidad de arribos son de mayo a octubre.

V. Lead time

```python
train['lead_time'].describe()
```

```python
v_lead_time_hist= snb.histplot(data=train['lead_time'])
v_lead_time_hist.set_title('Cantidad de días desde la fecha de reserva a la de llegada')
v_lead_time_hist.set_xlabel('Días')
v_lead_time_hist.set_ylabel('Frecuencia')                    
```

```python
density_lead_time=snb.kdeplot(data=train['lead_time']).set(title="Densidad lead time",xlabel="Dias",ylabel="Densidad")
```

```python
v_lead_time_box=snb.boxplot(data=train['lead_time']).set_title('Cantidad de días desde la fecha de reserva a la de llegada')
```

En promedio, las reservas se realizan con 112 días de anticipación a la fecha de arribo. Pero la mayor cantidad de reservas se realizan en el mismo día de llegada y con un maximo de 629 días.
