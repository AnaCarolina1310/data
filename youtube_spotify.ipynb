```python
df.drop([ 'Unnamed: 0', 'Url_spotify',  'Album_type',
       'Uri',  'Energy', 'Key', 'Loudness', 'Speechiness',
       'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 
        'Url_youtube',  'Channel', 'Description', 'Licensed',  'Stream', 'Tempo'], axis=1, inplace=True)
```

```python
df.dropna(subset = ['Artist','Track','Album',
       'Duration_ms','Views', 'Likes',
       'Comments', 'official_video', 'Danceability'], inplace = True)
```

```python
df.rename(columns = {'Danceability':'Danceability %', 'Duration_ms': 'Duration_mm', 'official_video': 'Video'}, inplace = True)
```

```python
df.shape
```

```python
df['Danceability %']= df['Danceability %'].apply(lambda x: int(x*100))
```

```python
df['Duration_mm']=df['Duration_mm'].apply(lambda x: round((x/1000)/60,2))
```

```python
df.loc[df['Video'] == True, 'Video']= 'Oficial'
df.loc[df['Video'] == False, 'Video']= 'No Oficial'
```

## Top ten: los 10 artistas más reproducidos

# Nivel de bailabilidad más reproducido


```python
df['Danceability level'] ='No dancebility level'
df.loc[(df['Danceability %'] > 0)& (df['Danceability %'] <= 33), 'Danceability level'] = 'Low'
df.loc[(df['Danceability %'] > 33)& (df['Danceability %'] <= 66), 'Danceability level'] = 'Medium'
df.loc[(df['Danceability %'] > 66), 'Danceability level'] = 'High'
```

```python
danceability_views= df.groupby('Danceability level')[['Views']].sum()
danceability_views.plot(kind='pie', y='Views')
```

Las canciones bailables son las más reproducidas.

# Comportamiento de los likes con respecto a las reproducciones 

```python
views_likes=df.groupby('Artist')[['Views', 'Likes']].sum().sort_values(by = ['Views'],ascending = False).reset_index()

import statsmodels.api as sm

y = views_likes['Likes']
x =views_likes['Views']

x=sm.add_constant(x)
lm=sm.OLS(y,x).fit()

lm.summary()
y_pred= 0.0062*x['Views'] +7.543

plt.figure(figsize=(6,4), tight_layout=True)
sns.scatterplot(x=x['Views'], y=y)

sns.lineplot(x=x['Views'], y=y_pred, color='red')

plt.xlim(0)
plt.ylim(0)
plt.savefig('afadf')
plt.show()
```

Comprobamos que el comportamiento de los likes con respecto a la cantidad de reproducciones es ascendente. Por lo que cuantas más reproducciones tenga una canción más probabilidades de ser likeada tiene.
